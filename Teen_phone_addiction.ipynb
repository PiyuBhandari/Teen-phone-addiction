```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
```

# Import Dataset

```python
df = np.loadtxt(r"D:\teen_phone_addiction_dataset.csv", delimiter=",", dtype=str)
# Convert it into a DataFrame
df = pd.DataFrame(df[1:], columns=df[0])
print(df.head())
```

# Clean Data

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values)

# Check for duplicate rows
duplicates = df.duplicated().sum()
print("\nNumber of duplicate rows:", duplicates)

# Clean the data by removing duplicates and rows with missing values
df_cleaned = df.drop_duplicates()
df_cleaned = df_cleaned.dropna()

# Show shape before and after cleaning
print("\nOriginal shape:", df.shape)
print("Cleaned shape:", df_cleaned.shape)
```

# Top 10 data

```python
Top_10_data=df[:10]
print(Top_10_data)
```

# Daily_Usage_Hours

```python
Mean=df["Daily_Usage_Hours"].astype(float).mean()
print("Mean of Daily Usage Hours : ",Mean)

Median=df["Daily_Usage_Hours"].astype(float).median()
print("Mdian of Daily Usage Hours : ",Median)

Mode=df["Daily_Usage_Hours"].astype(float).mode()
print("Mode of Daily Usage Hours : ",Mode)

std=df["Daily_Usage_Hours"].astype(float).std()
print("Standard Deviation of Daily Usage Hours : ",std)

var=df["Daily_Usage_Hours"].astype(float).var()
print("Variance of Daily Usage Hours : ",var)
```

# Weekend_Usage_Hours

```python
Mean=df["Weekend_Usage_Hours"].astype(float).mean()
print("Mean of Weekend Usage Hours : ",Mean)

Median=df["Weekend_Usage_Hours"].astype(float).median()
print("Mdian of Weekend Usage Hours : ",Median)

Mode=df["Weekend_Usage_Hours"].astype(float).mode()
print("Mode of Weekend Usage Hours : ",Mode)

std=df["Weekend_Usage_Hours"].astype(float).std()
print("Standard Deviation of Weekend Usage Hours : ",std)

var=df["Weekend_Usage_Hours"].astype(float).var()
print("Variance of Weekend Usage Hours : ",var)
```

# Slicing

```python
df_slice = df.loc[:10, ["Name", "Weekend_Usage_Hours", "Addiction_Level"]]
print(df_slice)
```

```python
# Teens who use their phones more than 6 hours on weekends
high_usage = df[df["Weekend_Usage_Hours"].astype(float) > 6]
print(high_usage)
```

# Line Chart

```python
df["Age"] = df["Age"].astype(int)
df["Weekend_Usage_Hours"] = df["Weekend_Usage_Hours"].astype(float)

plt.plot(df["Age"], df["Weekend_Usage_Hours"], marker="o")
plt.title("Weekend Usage Hours by Age")
plt.xlabel("Age")
plt.ylabel("Weekend Usage Hours")
plt.grid(True)
plt.show()
```

# Bar Chart

```python
gender_avg = df_cleaned.groupby("Gender")["Addiction_Level"].mean()
gender_avg.plot(kind="bar", color=["skyblue", "pink", "lightgreen"])
plt.title("Average Addiction Level by Gender")
plt.xlabel("Gender")
plt.ylabel("Addiction Level")
plt.show()
```

# Histogram

```python
df["Daily_Usage_Hours"].astype(float).plot(kind="hist", bins=10, color="violet", edgecolor="black")
plt.title("Distribution of Daily Usage Hours")
plt.xlabel("Hours")
plt.ylabel("Frequency")
plt.show()
```

# Box Plot

```python
# Step 1: Filter out malformed rows
valid_rows = df[df["Addiction_Level"].str.match(r'^\d+(\.\d+)?$')]

# Step 2: Convert to float
valid_rows["Addiction_Level"] = valid_rows["Addiction_Level"].astype(float)

# Step 3: Plot the boxplot again
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x="School_Grade", y="Addiction_Level", data=valid_rows)
plt.title("Addiction Level Across School Grades")
plt.xlabel("Grade")
plt.ylabel("Addiction Level")
plt.show()

```

# Scatter Plot

```python
plt.scatter(
    df["Daily_Usage_Hours"].astype(float),
    df["Academic_Performance"].astype(float),
    color="teal"
)
plt.title("Daily Usage vs Academic Performance")
plt.xlabel("Daily Usage Hours")
plt.ylabel("Academic Performance")
plt.show()
```

```python
#Convert numeric columns
numeric_df = df[[
    "Daily_Usage_Hours", "Weekend_Usage_Hours", "Sleep_Hours",
    "Addiction_Level", "Academic_Performance", "Phone_Checks_Per_Day"
]].astype(float)

#Calculate correlation matrix
correlation_matrix = numeric_df.corr()
print(correlation_matrix)

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

```python
purpose_counts = df["Phone_Usage_Purpose"].value_counts()

plt.pie(purpose_counts, labels=purpose_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Primary Phone Usage Purpose")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Convert the Age column to integers
df["Age"] = df["Age"].astype(int)

# Count occurrences of each age
age_counts = df["Age"].value_counts().sort_index()

# Create the pie chart
plt.figure(figsize=(4,4))
plt.pie(age_counts, labels=age_counts.index, autopct='%1.1f%%', startangle=140, 
        colors=plt.cm.Pastel1.colors)
plt.title("Age Distribution of Teens in Dataset")
plt.axis("equal")
plt.show()
```
